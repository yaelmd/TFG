```python
import pandas as pd
```

```python
task = 'epistemic_reasoning'
benchmark = 'bigbench'
df = pd.read_csv(f'./{benchmark}/{task}/gpt-4_{task}.csv',delimiter=";")
df
```

```python
# convert all the text to numeric values

for col in df.columns[4:]:
    errors=True
    while errors:
        try:
            df[col] = pd.to_numeric(df[col], errors='raise')
            errors=False
        except Exception as e:
            msg = str(e)
            print(msg)
            row_idx = int(msg.split(" ")[-1].split(")")[0])
            col_idx = df.columns.get_loc(col)
            print(row_idx)
            print(col_idx)
            # correct the value at the specified row and column indices
            bad_value = df.iloc[row_idx, col_idx].split(" ")
            print(bad_value)
            if bad_value[0]=='N/A':
                corrected_value = 0
            elif bad_value[0]=='Infinite':
                corrected_value = 5
            #Very concrete cases are forced to zero (done like this to take control of the different cases)
            #Can be optimized
            elif bad_value[0]=='Sentence:':
                corrected_value = 0
            elif bad_value[0]=='"I':
                corrected_value = 0
            elif bad_value[0]=='\'Nobody':
                corrected_value = 0
            elif bad_value[0]=='"The':
                corrected_value = 0
            elif bad_value[0]=='The':
                corrected_value = 0
            elif bad_value[0] == '"It':
                corrected_value = 0
            elif bad_value[0] == '"John':
                corrected_value = 0
            elif bad_value[0] == 'Invalid':
                corrected_value = 0
            elif bad_value[0] == 'Not':
                corrected_value = 0
            elif bad_value[0] == 'Please':
                corrected_value = 0
            elif bad_value[0] == 'This':
                corrected_value = 0
            elif bad_value[0] == '\'He':
                corrected_value = 0
            elif bad_value[0] == '"He':
                corrected_value = 0
            elif bad_value[0] == '-1':
                corrected_value = 0
            elif bad_value[0] == 'Incomplete':
                corrected_value = 0
            elif bad_value[0] == 'Unclear':
                corrected_value = 0
            elif bad_value[0] == 'C.' or bad_value[0] == 'C':
                corrected_value = 0
            elif bad_value[0] == 'B.' or bad_value[0] == 'B':
                corrected_value = 0
            elif bad_value[0] == 'A.' or bad_value[0] == 'A':
                corrected_value = 0
            elif '/' in bad_value[0]:
                bad_value = bad_value[0].split('/')
                corrected_value = int(bad_value[0])/int(bad_value[1])
            else:
                corrected_value = int(bad_value[0][0])
            print(corrected_value)
         # replace this with your correction logic
            df.iloc[row_idx, col_idx] = corrected_value

last_columns = df.columns[4:]

df[last_columns] = df[last_columns].applymap(lambda x: x if x >= 0 else 0)
#Fill all the NaN columns
df = df.fillna(0)
```

```python
df.to_csv(f'./{benchmark}/{task}/{task}_features_post.csv',index=False)
df
```
